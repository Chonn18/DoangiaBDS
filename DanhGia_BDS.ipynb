MODEL NEURAL NETWORK

```python
def get_model():

    model = Sequential([
        Dense(30, input_shape = (8,16), activation = 'relu'), #30 neurons, Input Layer, 'relu' activation
        Dense(30, activation = 'relu'),                      #30 neurons, Input Layer, 'relu' activation
        Dense(30, activation = 'relu'),                      #30 neurons, Input Layer, 'relu' activation
        Dense(30, activation = 'relu'),                      #30 neurons, Input Layer, 'relu' activation
        Dense(1, activation = 'linear')

    ])
    adam = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(
        loss='mse', # using mse loss function
        metrics = [tf.keras.metrics.RootMeanSquaredError()],
        optimizer ='adam'
        )
    return model
model = get_model()
model.summary()
```

TRAIN MODEL NEURAL NETWORK

```python
model = get_model()
history = model.fit(
    x_train, y_train,
    batch_size=8,
    validation_data = (x_Validation, y_Validation),
    epochs = 1000
)
```
